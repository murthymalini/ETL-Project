# COIN API EXTRACT FROM CRYPTO COMPARE

```python
# list of coins to be used in our project

PriceTable = {}
price_API_query = pd.DataFrame()

coins = ["BTC", "ETH", "LTC", "BCH","EOS", "BNB", "XLM", "TRX", "ADA", "NEO" ]
# coin = ('BTC')

limit = 859  # verify the number of days you wish to query

for coin in coins:
    PriceTable[coin] = []
    url =  f"https://min-api.cryptocompare.com/data/v2/histoday?fsym={coin}&tsym=USD&limit={limit}&api_key={api_key}"
    response = requests.get(url)
    data = response.json()
    PriceTable[coin].append(data)
    PriceTable
    Price2 = PriceTable[coin][0]['Data']['Data']
    price_df = pd.DataFrame(Price2)
    price_df = price_df['close']
    price_API_query[coin] = price_df
```
