# EXCHANGE VOLUMEN EXTRACT FROM CRYPTO COMPARE

```python
# list of coins to be used in our project

Exch_Vol = {}
exch_vol_API_query = pd.DataFrame()

coins = ["BTC", "ETH", "LTC", "BCH","EOS", "BNB", "XLM", "TRX", "ADA", "NEO" ]
# coins = ['BTC', "ETH"] # test sample

limit = 859 # verify the number of days you wish to query

for coin in coins:
    Exch_Vol[coin] = []
    url = f"https://min-api.cryptocompare.com/data/exchange/symbol/histoday?fsym={coin}&tsym=USD&limit={limit}&api_key={api_key}"
    response = requests.get(url)
    data = response.json()
    Exch_Vol[coin].append(data)
    Exch_Vol
    VT = Exch_Vol[coin][0]['Data']
    exch_vol_df = pd.DataFrame(VT)
    exch_vol_df = exch_vol_df['volumetotal']
    exch_vol_API_query[coin] = exch_vol_df
```
