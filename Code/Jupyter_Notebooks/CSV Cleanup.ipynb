```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
# Create a reference the CSV file desired
csv_path = "crd.csv"

# Read the CSV into a Pandas DataFrame
crd_df = pd.read_csv(csv_path)

# Print the first five rows of data to the screen
crd_df.head()
```

```python
# reset index
crd_df.reset_index()
```

```python
# export to csv 
crd_df.to_csv("clean_crd_df.csv")
```

```python
# Filter the data so that only date for the last 2 years are in a DataFrame
clean_crd_df = crd_df.drop(crd_df.index[0:342])
```

```python
# select only rows needed for analysis
clean_crd_df = clean_crd_df[["date", "CapMrktCurUSD", "PriceUSD"]]
clean_crd_df.head()
```

```python
# add number of coins column
organized_crd_number = clean_crd_df['CapMrktCurUSD'].div(clean_crd_df['PriceUSD'])
clean_crd_df['number_of_coins'] = organized_crd_number
clean_crd_df.head()
```

```python
# Rename headers
clean_crd_df = clean_crd_df.rename(
    columns={"date": "Date", "CapMrktCurUSD": "Market Cap (USD)", "PriceUSD": "Price (USD)", "number_of_coins": "Number of Coins"})
clean_crd_df.head()
```
