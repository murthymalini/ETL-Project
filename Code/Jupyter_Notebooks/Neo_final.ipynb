```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
csv_path = "neo.csv"
neo_df = pd.read_csv(csv_path)
neo_df.head()
```

```python
clean_neo_df = neo_df.dropna(how="any")
clean_neo_df.head()
```

```python
clean_neo_df.reset_index()
```

```python
# export to csv 
clean_neo_df.to_csv("clean_neo_df.csv")
```

```python
clean_neo_df = clean_neo_df.drop(clean_neo_df.index[384:479])
```

```python
organized_neo_df = clean_neo_df[["date", "CapMrktCurUSD", "PriceUSD"]]
organized_neo_df.head()
```

```python
organized_neo_number = organized_neo_df['CapMrktCurUSD'].div(organized_neo_df['PriceUSD'])
organized_neo_df['number_of_coins'] = organized_neo_number
organized_neo_df.head()
```

```python
organized_neo_df = organized_neo_df.rename(
    columns={"date": "Date", "marketcap_usd": "Market Cap (USD)", "price_usd": "Price (USD)", "number_of_coins": "Number of Coins"})
organized_neo_df.head()
```

```python
count = organized_neo_df["Date"].count()
print(count)
```

```python
organized_neo_df.to_csv("organized_neo_df.csv")
```
