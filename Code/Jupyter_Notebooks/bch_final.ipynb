```python
import pandas as pd
```

```python
csv_path = "bch.csv"
bch_df = pd.read_csv(csv_path)
bch_df.head()
```

```python
clean_bch_df = bch_df.dropna(how="any")
clean_bch_df.head()
```

```python
clean_bch_df.reset_index()
```

```python
# export to csv 
clean_bch_df.to_csv("clean_bch_df.csv")
```

```python
clean_bch_df = clean_bch_df.drop(clean_bch_df.index[336:461])
```

```python
organized_bch_df = clean_bch_df[["date", "CapMrktCurUSD", "PriceUSD"]]
organized_bch_df.head()
```

```python
organized_bch_number = organized_bch_df['CapMrktCurUSD'].div(organized_bch_df['PriceUSD'])
organized_bch_df['number_of_coins'] = organized_bch_number
organized_bch_df.head()
```

```python
organized_bch_df = organized_bch_df.rename(
    columns={"date": "Date", "marketcap_usd": "Market Cap (USD)", "price_usd": "Price (USD)", "number_of_coins": "Number of Coins"})
organized_bch_df.head()
```

```python
count = organized_bch_df["Date"].count()
print(count)
```
