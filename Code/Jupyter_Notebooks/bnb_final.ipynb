```python
bnb_df = bnb_df.drop(bnb_df.index[107:202])
```

```python
organized_bnb_df = bnb_df[["date", "CapMrktCurUSD", "PriceUSD"]]
organized_bnb_df.head()
```

```python
organized_bnb_number = organized_bnb_df['CapMrktCurUSD'].div(organized_bnb_df['PriceUSD'])
organized_bnb_df['number_of_coins']= organized_bnb_number
organized_bnb_df.head()
```

```python
organized_bnb_df = organized_bnb_df.rename(
    columns={"date": "Date", "marketcap_usd": "Market Cap (USD)", "price_usd": "Price (USD)", "number_of_coins": "Number of Coins"})
organized_bnb_df.head()
```
